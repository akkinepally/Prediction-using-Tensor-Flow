```python
import pandas as pd 
import tensorflow as tf
```

```python
#loading the training data
train = pd.read_csv('sales_data_training.csv')
```

```python
train
```

```python
#Separating dependent and independent features
x_train = train.iloc[:, 0:9].values
y_train = train.iloc[:, [9]].values
```

```python
#load test data 
test = pd.read_csv('sales_data_test.csv')
```

```python
# seperating depedent and indepent features
x_test = test.iloc[:, 0:9].values
y_test = test.iloc[:, [9]].values
```

```python
# scaling the data 
from sklearn.preprocessing import MinMaxScaler
x_scaler = MinMaxScaler(feature_range = (0,1))
y_scaler = MinMaxScaler(feature_range = (0,1))
```

```python
# scale training data 
x_train_scaler = x_scaler.fit_transform(x_train)
y_train_scaler = y_scaler.fit_transform(y_train)
```

```python
#scale test data 
x_test_scaler = x_scaler.transform(x_test)
y_test_scaler = y_scaler.transform(y_test)
```

```python
# Hyper Parameter 
learning_rate = 0.001
training_epochs = 100
display_step = 5
```

```python
# defining number of inputs and outputs 
number_of_inputs = 9
number_of_outputs = 1
```

```python
# defining number of nuerons and number of layers 
layer_1 = 50 
layer_2 = 100
layer_3 = 50 
```

```python
# creating scope for each layer(input) 
with tf.variable_scope('input'):
     x = tf.placeholder(tf.float32, shape =(None, number_of_inputs))
    
```

```python

    # scope for layer 1
with tf.variable_scope('layer1'):
    weights = tf.get_variable(name = 'weights1', shape = [number_of_inputs, layer_1], initializer = tf.contrib.layers.xavier_initializer())
    biases = tf.get_variable(name = 'biases1', shape = [layer_1], initializer = tf.zeros_initializer())
    layer1_output = tf.nn.relu(tf.matmul(x, weights) + biases)
```

```python
# scope for layer 2
with tf.variable_scope('layer2'):
    weights = tf.get_variable(name = 'weights2', shape = [layer_1, layer_2], initializer = tf.contrib.layers.xavier_initializer())
    biases = tf.get_variable(name = 'biases2', shape = [layer_2], initializer = tf.zeros_initializer())
    layer2_output = tf.nn.relu(tf.matmul(layer1_output, weights) + biases)
```

```python
# scope for layer 3
with tf.variable_scope('layer3'):
    weights = tf.get_variable(name ='weights3', shape = [layer_2, layer_3], initializer = tf.contrib.layers.xavier_initializer())
    biases = tf.get_variable(name = 'biases3', shape = [layer_3], initializer = tf.zeros_initializer())
    layer3_output  = tf.nn.relu(tf.matmul(layer2_output, weights) + biases)
    
```

```python
# scope for output
with tf.variable_scope('output'):
    weights = tf.get_variable(name = 'weights4', shape = [layer_3, number_of_outputs], initializer = tf.contrib.layers.xavier_initializer())
    biases = tf.get_variable(name = 'biases4', shape = [number_of_outputs], initializer = tf.zeros_initializer())
    prediction = tf.nn.relu(tf.matmul(layer3_output, weights) + biases)
```

```python
# scope for cost function 
with tf.variable_scope('cost'):
    y = tf.placeholder(tf.float32, shape = (None, 1))
    cost = tf.reduce_mean(tf.squared_difference(prediction, y))
```

```python
# optimizer 
with tf.variable_scope('train'):
    optimizer = tf.train.AdamOptimizer(learning_rate).minimize(cost)
```

```python
# session 
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)
```

```python
# training loop 
for epoch in range(training_epochs):
    sess.run(optimizer, feed_dict= {x : x_train_scaler, y : y_train_scaler})
    if epoch%5 == 0:
        training_cost = sess.run(cost, feed_dict= {x : x_train_scaler, y: y_train_scaler})
        testing_cost = sess.run(cost, feed_dict= {x : x_test_scaler, y: y_test_scaler})
        print(epoch, training_cost, testing_cost)
print('Training is complete')
final_training_cost = sess.run(cost, feed_dict= {x : x_train_scaler, y: y_train_scaler})
final_testing_cost = sess.run(cost, feed_dict= {x : x_test_scaler, y: y_test_scaler})

print('final_training_cost: {}'.format(final_training_cost))
print('final_testing_cost: {}'.format(final_testing_cost))

   
```
